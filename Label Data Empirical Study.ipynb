```python
import pandas as pd
```

```python
data = pd.read_csv("reports_data.csv",index_col=0)
data
```

```python


# Refine criteria for human-detectable vulnerabilities
human_criteria = ['complex', 'nuanced', 'interpretation', 'subjective', 'context-dependent', 'difficult', 'experience', 'challenging', 'judgment', 'sensitive']

# Refine criteria for machine-detectable vulnerabilities
machine_criteria = ['syntax', 'security', 'automated', 'static', 'analysis', 'code', 'known', 'automatically', 'logical', 'algorithm', 'predictable', 'consistent']

# Expand the lists of keywords
human_keywords = ['complex', 'nuanced', 'interpretation', 'subjective', 'context-dependent', 'difficult', 'experience', 'challenging', 'judgment', 'sensitive', 'additional', 'keywords', 'for', 'human', 'detectable']
machine_keywords = ['syntax', 'security', 'automated', 'static', 'analysis', 'code', 'known', 'automatically', 'logical', 'algorithm', 'predictable', 'consistent', 'additional', 'keywords', 'for', 'machine', 'detectable']

# Weighted scoring
def weighted_keyword_score(text, keyword_weights):
    return sum(keyword_weights.get(word, 1) * (word in text) for word in keyword_weights)

# Stricter weights for human-detectable vulnerabilities
human_weights = {'complex': 2, 'nuanced': 2, 'interpretation': 2, 'subjective': 2, 'context-dependent': 2, 'difficult': 2, 'experience': 2, 'challenging': 2, 'judgment': 2, 'sensitive': 2}

# Stricter weights for machine-detectable vulnerabilities
machine_weights = {'syntax': 2, 'security': 2, 'automated': 2, 'static': 2, 'analysis': 2, 'code': 2, 'known': 2, 'automatically': 2, 'logical': 2, 'algorithm': 2, 'predictable': 2, 'consistent': 2}

# Function to label smart contract reports based on refined criteria and stricter weights
def label_vulnerabilities(text):
    human_score = weighted_keyword_score(text, human_weights)
    machine_score = weighted_keyword_score(text, machine_weights)
    
    # Stricter threshold for labeling
    threshold = 2.0
    
    if human_score > threshold:
        return 'Human-Detectable'
    elif machine_score > threshold:
        return 'Machine-Detectable'
    else:
        return 'Uncertain'

# Apply the labeling function to create a new 'label' column
data['Label'] = data['Report_Content'].apply(label_vulnerabilities)

# Display counts of each label
print(data['Label'].value_counts())
```
